# View All Images in generated_fixed

This notebook loads and displays all images in `generated_fixed/` matching the pattern `fixed-*.png`.


```python
import math
import pathlib

import matplotlib.pyplot as plt
from PIL import Image
```

```python
GENERATED_DIR = pathlib.Path('generated_fixed')
IMAGE_PATHS = sorted(GENERATED_DIR.glob('fixed-*.png'))

print('generated_fixed exists:', GENERATED_DIR.exists())
print('num images found:', len(IMAGE_PATHS))

if IMAGE_PATHS:
    print('first image:', IMAGE_PATHS[0].name)
    print('last image:', IMAGE_PATHS[-1].name)
else:
    print('No images found. Expected files like fixed-001.png in generated_fixed/.')
```

```python
def show_all_images(image_paths, images_per_row=5, figsize_per_image=3.0):
    if not image_paths:
        raise FileNotFoundError('No fixed-*.png files found in generated_fixed/.')

    num_images = len(image_paths)
    num_rows = math.ceil(num_images / images_per_row)
    fig, axes = plt.subplots(
        num_rows,
        images_per_row,
        figsize=(images_per_row * figsize_per_image, num_rows * figsize_per_image)
    )

    if num_rows == 1:
        axes = [axes] if images_per_row == 1 else axes

    axes = list(axes.ravel()) if hasattr(axes, 'ravel') else list(axes)

    for ax, image_path in zip(axes, image_paths):
        ax.imshow(Image.open(image_path))
        ax.set_title(image_path.stem)
        ax.axis('off')

    for ax in axes[len(image_paths):]:
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
show_all_images(IMAGE_PATHS, images_per_row=5, figsize_per_image=3.0)
```
